```python
import keras
import tensorflow as tf
from keras.models import Model
from keras.layers import Dense
from keras import optimizers
from keras.preprocessing.image import ImageDataGenerator
from keras.preprocessing import image
```

```python
train_datagen = ImageDataGenerator(rescale=1./255,
    shear_range=0.2,
    zoom_range=0.2,
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    horizontal_flip=True,
    validation_split=0.33) # set validation split

train_generator = train_datagen.flow_from_directory(
    labeled_path,
    target_size=(224,224),
    batch_size=32,
    class_mode='binary',
    subset='training'
)

validation_generator = train_datagen.flow_from_directory(
    labeled_path, # same directory as training data
    target_size=(224,224),
    batch_size=32,
    class_mode='binary',
    subset='validation'
)
```

```python
from keras.applications.vgg16 import VGG16
vggmodel = VGG16(weights='imagenet', include_top=True)
vggmodel_1 = VGG16(weights='imagenet', include_top=False)
```

```python
vggmodel.summary()
```

```python
vggmodel_1.summary()
```

```python
for layers in (vggmodel.layers)[:19]:
    print(layers)
    layers.trainable = False
```

```python
vggmodel.summary()
```

```python
# Add a layer where input is the output of the second last layer
num_classes = 2
predictions = Dense(num_classes, activation="softmax", name='output_predictions')(vggmodel.layers[-2].output)
```

```python
model_final = Model(inputs = vggmodel.input, outputs = predictions)
```

```python
model_final.compile(loss = "sparse_categorical_crossentropy",
                    optimizer = tf.keras.optimizers.SGD(learning_rate=0.0001, momentum=0.9), 
                    metrics=["accuracy"])
```

```python
model_final.summary()
```

```python
from keras.callbacks import ModelCheckpoint, EarlyStopping
checkpoint = ModelCheckpoint("vgg16_1.h5", 
                             monitor='val_accuracy', 
                             verbose=1, 
                             save_best_only=True,
                             save_weights_only=True, 
                             mode='max', 
                             save_freq=1)

early = EarlyStopping(monitor='val_accuracy', 
                      min_delta=0, 
                      patience=2, 
                      verbose=1, 
                      mode='max')

history = model_final.fit(
                train_generator, 
                epochs= 10, 
                validation_data= validation_generator, 
                validation_freq=1, 
                callbacks=[checkpoint,early])
```

```python
model_final.save_weights("vgg16_1.h5")
```

```python
import matplotlib.pyplot as plt
plt.plot(history.history["accuracy"])
plt.plot(history.history['val_accuracy'])
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title("model accuracy")
plt.ylabel("Accuracy")
plt.xlabel("Epoch")
plt.legend(["Accuracy","Validation Accuracy","loss","Validation Loss"])
plt.show()
```
